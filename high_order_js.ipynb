```python
%%writefile code.js

// A
const data = [5, 12, 3, 20, 11, 9, 30];
const newData = data.filter(isAboveTen).map(double)
const expectedOutputA = [24, 40, 22, 60]


function isAboveTen (num) {
  return num > 10
}

function double (num) {
  return num * 2
}

console.log("")
console.log("== A ==")
console.log("My output: " + newData)
console.log("Expected output: " + expectedOutputA)

// B

const numbers = [1, 2, 3, 4, 5, 6];
const expectedOutputB = 24

function isEven (num) {
  return num % 2 === 0
}

const evenNumbersDoubled = numbers.filter(isEven).map(double)

let sum = 0;
for (let i of evenNumbersDoubled) {
  sum += i
}

console.log("")
console.log("== B ==")
console.log("My output: " + sum)
console.log("Expected output: " + expectedOutputB)

// C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

let under18 = users.some(user => user.age < 18)

function idIsThree(user) {
  return user.id === 3
}

let user3 = users.find(idIsThree)

let usersOutput = ""
users.forEach(user => {
  usersOutput += user.name + "\n"
  })

console.log("")
console.log("== C ==")
console.log("Someone is under 18?: " + under18 )
console.log("Find user with id 3: " + JSON.stringify(user3))
console.log("Users: \n" + usersOutput )


// D
function repeat(n, action) {
  for (let i = 0; i < n - 1; i++) {
    action(i);
  }
}

console.log("")
console.log("== D ==")
repeat(3, console.log) // 0, 1, 2
repeat(4, i => console.log("Step " + i))






```
